```python
from matplotlib import pyplot as plt

a = [6,5,4,7,9,6,1,5]
```

```python
x = [5,6,9,8,5,6,5,4]
y = [3,6,5,4,8,5,4,9]
```

```python
plt.scatter(a,x, marker='.', c='green',s=30)
plt.scatter(a,y,marker = '.', c= 'red', s=20)
plt.show()
```

## subplots

```python
plt.subplot(1,2,1)
plt.scatter(a,x, marker='.', c='green',s=30)


plt.subplot(1,2,2)
plt.scatter(a,y,marker = '.', c= 'red', s=20)

plt.show()
```
