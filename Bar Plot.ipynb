```python
from matplotlib import pyplot as plt
import numpy as np

student = {
    'Sugat' : 56,
    'Manish': 44,
    'Sameer': 99,
    'Sagar' : 73,
 'Shaubham' : 24,
    'Ayush' : 99,
}
student
```

```python
Names = list(student.values())
Marks = list(student.keys())

plt.bar(Marks,Names)
plt.show()
```

## In details bar graph

```python
Names = list(student.values())
Marks = list(student.keys())

plt.bar(Marks,Names)
plt.show()
```

```python
Names = list(student.values())
Marks = list(student.keys())

plt.title('RESULT')
plt.xlabel('Names')
plt.ylabel('Marks')

plt.grid(True)

plt.bar(Marks,Names)
plt.show()
```

## Horizontal Graph with color

```python
Names = list(student.values())
Marks = list(student.keys())

plt.title('RESULT')
plt.ylabel('Names')
plt.xlabel('Marks')

plt.grid(True)

plt.barh(Marks,Names, color='y')
plt.show()
```
