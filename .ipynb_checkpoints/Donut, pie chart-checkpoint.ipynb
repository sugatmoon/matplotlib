```python
from matplotlib import pyplot as plt
```

```python
x = [ 'Sugat', 'Manish', 'Sameer', 'Shubham', 'Pranay','Aniket','Swaraj']
m = [65,99,92,44,77,68,93]
```

# show the pie

```python
plt.pie(m,labels = x)
plt.show()
```

# with percentage

```python
plt.pie(m,labels = x,autopct='%0.1f%%',colors=['black','green','white','yellow','red','Blue','Purple'])
plt.show()
```

# Donut

```python
plt.pie(m,labels = x,autopct='%0.1f%%', radius=5)
plt.pie([5],colors=['yellow'],radius=2)
plt.show()
```
