# Single Line graph

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
a1 = np.arange(1,24)
a1 = a1*3
```

```python
a1
```

```python
a2 = a1 * 2
```

```python
a1
```

```python
a2
```

## Show curve of graph

```python
plt.plot(a1,a2)
plt.show()
```

## Title label of x and  y axis

```python
plt.plot(a1,a2)
plt.title('single line curve')
plt.xlabel('Petrol ----->')
plt.ylabel('Deseal ----->')
plt.show()
```

## Change the color, style, width of curve line

```python
plt.plot(a1,a2, color = 'r', linestyle = ':', linewidth = '3')
plt.title('single line curve')
plt.xlabel('Petrol ----->')
plt.ylabel('Deseal ----->')
plt.show()
```

```python
plt.plot(a1,a2)
plt.title('single line curve')
plt.xlabel('Petrol ----->')
plt.ylabel('Deseal ----->')
plt.show()
```

## Multiple curve line, show grid in a graph

```python
b1 = 6 * a1
b2 = 11 * a1
```

```python
b1
```

```python
b2
```

```python
plt.plot(a1 , b1, color = 'g', linestyle = ':' ,linewidth = '5')
plt.plot(a2 , b2, color = 'r',linestyle = '-.' ,linewidth = '6')
plt.ylabel('Green +++++++++++++++>')
plt.xlabel('Yellow------------->')
plt.grid(True)
plt.show()
```

## Sub graph

```python
plt.subplot(2,1,1)
plt.plot(a1 , b1, color = 'g' )
plt.subplot(2,1,2)
plt.plot(a2 , b2, color = 'r')
plt.show()
```
